```python
# Mapas de tradução
map_educacao = {
    1: 'Abaixo da faculdade',
    2: 'Faculdade',
    3: 'Bacharelado',
    4: 'Mestrado',
    5: 'Doutorado'
}

map_satisfacao = {
    1: 'Baixo',
    2: 'Médio',
    3: 'Alto',
    4: 'Muito alto'
}

map_envio = {
    1: 'Baixo',
    2: 'Médio',
    3: 'Alto',
    4: 'Muito alto'
}

map_desempenho = {
    1: 'Baixo',
    2: 'Bom',
    3: 'Excelente',
    4: 'Excelente'
}

map_relacionamento = {
    1: 'Baixo',
    2: 'Médio',
    3: 'Alto',
    4: 'Muito alto'
}

map_vida_trabalho = {
    1: 'Ruim',
    2: 'Bom',
    3: 'Melhor',
    4: 'Melhor'
}

# Aplicar mapeamentos
df['nivel_educacional'] = df['nivel_educacional'].map(map_educacao)
df['satisfacao_ambiente'] = df['satisfacao_ambiente'].map(map_satisfacao)
df['envolvimento_trabalho'] = df['envolvimento_trabalho'].map(map_envio)
df['satisfacao_trabalho'] = df['satisfacao_trabalho'].map(map_satisfacao)
df['avaliacao_desempenho'] = df['avaliacao_desempenho'].map(map_desempenho)
df['satisfacao_relacionamento'] = df['satisfacao_relacionamento'].map(map_relacionamento)
df['equilibrio_vida_trabalho'] = df['equilibrio_vida_trabalho'].map(map_vida_trabalho)
```

```python
# Distância média de casa agrupada por cargo e desligamento
distancia_por_cargo_desligamento = df.groupby(['cargo', 'desligamento'])['distancia_de_casa'].mean().unstack()
distancia_por_cargo_desligamento
```

```python
# Salário médio mensal por educação e desligamento
renda_por_educacao_desligamento = df.groupby(['nivel_educacional', 'desligamento'])['salario_mensal'].mean().unstack()
renda_por_educacao_desligamento
```

Quanto maior o nível educacional, maior o salário médio, tanto para quem ficou quanto para quem saiu.

Pessoas com menor escolaridade têm menor salário e maior taxa de desligamento proporcionalmente.

A diferença entre os salários de quem ficou vs. quem saiu é marcante, sugerindo que:

Profissionais com menor salário têm maior propensão ao desligamento.

Retenção pode estar associada a melhores salários (mas também a outros fatores: satisfação, oportunidades, cargo, etc).

```python
# Taxa de desligamento por área ou cargo
df.groupby('cargo')['desligamento_flag'].mean().sort_values(ascending=False)
```

Representantes de Vendas têm a maior taxa de saída (quase 40%), indicando talvez pressão por metas, baixa remuneração, ou falta de perspectiva.

Técnicos de laboratório e RH também têm taxas elevadas, o que pode indicar problemas estruturais nessas áreas.

Cargos de liderança como Gerente e Diretor têm as menores taxas de desligamento, o que é comum em empresas, pois geralmente envolvem maior estabilidade, melhores salários e engajamento.



```python
# Distribuição da satisfação com o trabalho entre os que saíram ou não
df.groupby(['satisfacao_trabalho', 'desligamento']).size().unstack()
```

Funcionários com baixa satisfação têm uma taxa de desligamento quase o dobro dos que estão muito satisfeitos.

Mesmo os que se dizem com satisfação “Alta” ou “Média” ainda têm taxas relevantes (em torno de 16%).

A menor taxa de saída está entre quem declarou “muito alto” nível de satisfação.

Isso sugere fortemente que aumentar a satisfação no trabalho pode reduzir a rotatividade!

```python
#Balanceamento entre vida pessoal e profissional vs desligamento
df.groupby(['equilibrio_vida_trabalho', 'desligamento']).size().unstack()
```

Pessoas que classificaram o equilíbrio como "Ruim" têm uma taxa de desligamento mais que o dobro dos que consideraram "Melhor".

Mesmo entre os que consideraram “Bom”, a taxa ainda é elevada (quase 17%).

O resultado sugere fortemente que falta de equilíbrio entre vida pessoal e profissional está diretamente associada ao turnover.

```python
import pandas as pd
import os
import boto3


def executar_pipeline_etl(data_execucao):
    """
    Executa o pipeline ETL para dados de RH.

    Args:
        data_execucao (str): Data da execução para fins de log ou nome de arquivo.
    """
    print(f"[{data_execucao}] Iniciando execução do pipeline ETL...")

    # --- 1. EXTRAÇÃO DE DADOS ---
    print(f"[{data_execucao}] Etapa 1: Extração de Dados...")

    try:
        df = pd.read_csv('/content/drive/MyDrive/Data girls - Engenharia de Dados/Projeto_final_EngDados/WA_Fn-UseC_-HR-Employee-Attrition.csv')
        print(f"[{data_execucao}] Dados extraídos. Linhas: {df.shape[0]}, Colunas: {df.shape[1]}")
    except Exception as e:
        print(f"[{data_execucao}] Erro na extração: {e}")
        return # Aborta se a extração falhar

    # --- 2. TRANSFORMAÇÃO DE DADOS ---
    print(f"[{data_execucao}] Etapa 2: Transformação de Dados...")

    # Renomear colunas
    df.rename(columns={
        'Age': 'idade', 'Attrition': 'desligamento', 'BusinessTravel': 'viagem_a_trabalho',
        'DailyRate': 'taxa_diaria', 'Department': 'departamento', 'DistanceFromHome': 'distancia_casa_trabalho',
        'Education': 'educacao', 'EducationField': 'area_de_estudo', 'EmployeeCount': 'contagem_funcionarios',
        'EmployeeNumber': 'numero_funcionario', 'EnvironmentSatisfaction': 'satisfacao_ambiente',
        'Gender': 'genero', 'HourlyRate': 'taxa_horaria', 'JobInvolvement': 'envolvimento_trabalho',
        'JobLevel': 'nivel_cargo', 'JobRole': 'cargo', 'JobSatisfaction': 'satisfacao_cargo',
        'MaritalStatus': 'estado_civil', 'MonthlyIncome': 'renda_mensal', 'MonthlyRate': 'taxa_mensal',
        'NumCompaniesWorked': 'numero_empresas_anteriores', 'Over18': 'maior_18', 'OverTime': 'hora_extra',
        'PercentSalaryHike': 'aumento_salario_percentual', 'PerformanceRating': 'avaliacao_performance',
        'RelationshipSatisfaction': 'satisfacao_relacionamento', 'StandardHours': 'horas_padrao',
        'StockOptionLevel': 'nivel_opcao_acoes', 'TotalWorkingYears': 'total_anos_trabalhados',
        'TrainingTimesLastYear': 'treinamentos_ultimo_ano', 'WorkLifeBalance': 'equilibrio_vida_trabalho',
        'YearsAtCompany': 'anos_na_empresa', 'YearsInCurrentRole': 'anos_na_funcao_atual',
        'YearsSinceLastPromotion': 'anos_desde_ultima_promocao', 'YearsWithCurrManager': 'anos_com_gerente_atual'
    }, inplace=True)

    # Traduzir valores categóricos
    df['desligamento'] = df['desligamento'].map({'Yes': 'Sim', 'No': 'Não'})
    df['genero'] = df['genero'].map({'Male': 'Masculino', 'Female': 'Feminino'})
    df['estado_civil'] = df['estado_civil'].map({'Single': 'Solteiro(a)', 'Married': 'Casado(a)', 'Divorced': 'Divorciado(a)'})
    df['viagem_a_trabalho'] = df['viagem_a_trabalho'].map({'Travel_Rarely': 'Raramente', 'Travel_Frequently': 'Frequentemente', 'Non-Travel': 'Nao viaja'})
    df['area_de_estudo'] = df['area_de_estudo'].map({'Life Sciences': 'Ciencias da Vida', 'Medical': 'Medicina', 'Marketing': 'Marketing', 'Technical Degree': 'Tecnico', 'Other': 'Outros', 'Human Resources': 'Recursos Humanos'})
    df['hora_extra'] = df['hora_extra'].map({'Yes': 'Sim', 'No': 'Não'})
    df['cargo'] = df['cargo'].map({
        'Sales Executive': 'Executivo de Vendas', 'Research Scientist': 'Cientista Pesquisador',
        'Laboratory Technician': 'Tecnico de Laboratorio', 'Manufacturing Director': 'Diretor de Manufatura',
        'Healthcare Representative': 'Representante de Saude', 'Manager': 'Gerente',
        'Sales Representative': 'Representante de Vendas', 'Research Director': 'Diretor de Pesquisa',
        'Human Resources': 'Recursos Humanos'
    })
    df['educacao'] = df['educacao'].map({1: 'Abaixo da Media', 2: 'Media', 3: 'Acima da Media', 4: 'Excelente', 5: 'Otima'})
    df['satisfacao_ambiente'] = df['satisfacao_ambiente'].map({1: 'Baixa', 2: 'Media', 3: 'Alta', 4: 'Muito Alta'})
    df['envolvimento_trabalho'] = df['envolvimento_trabalho'].map({1: 'Baixo', 2: 'Medio', 3: 'Alto', 4: 'Muito Alto'})
    df['satisfacao_cargo'] = df['satisfacao_cargo'].map({1: 'Baixa', 2: 'Media', 3: 'Alta', 4: 'Muito Alta'})
    df['avaliacao_performance'] = df['avaliacao_performance'].map({1: 'Ruim', 2: 'Bom', 3: 'Otimo', 4: 'Excelente'}) # Correção: Geralmente 3 e 4
    df['satisfacao_relacionamento'] = df['satisfacao_relacionamento'].map({1: 'Baixa', 2: 'Media', 3: 'Alta', 4: 'Muito Alta'})
    df['equilibrio_vida_trabalho'] = df['equilibrio_vida_trabalho'].map({1: 'Ruim', 2: 'Bom', 3: 'Otimo', 4: 'Excelente'})

    # Criar faixa_etaria
    bins_idade = [18, 25, 35, 45, 55, 60]
    labels_idade = ['18-25', '26-35', '36-45', '46-55', '56+']
    df['faixa_etaria'] = pd.cut(df['idade'], bins=bins_idade, labels=labels_idade, right=False)

    # Criar desligamento_flag
    df['desligamento_flag'] = df['desligamento'].apply(lambda x: 1 if x == 'Sim' else 0)

    # Criar faixa_salarial
    df['faixa_salarial'] = pd.qcut(df['renda_mensal'], q=4, labels=['Q1_Baixa', 'Q2_Media', 'Q3_Alta', 'Q4_Muito_Alta'])

    # One-Hot Encoding
    categorical_cols = df.select_dtypes(include='object').columns
    df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True, dtype=int)

    # Tratamento de Nulos/Duplicatas (adicione aqui se fez ou verifique se não há)
    if df_encoded.isnull().sum().sum() > 0:
        print(f"[{data_execucao}] Alerta: Valores nulos encontrados após transformações.")
    if df_encoded.duplicated().sum() > 0:
        print(f"[{data_execucao}] Alerta: Registros duplicados encontrados. Removendo...")
        df_encoded.drop_duplicates(inplace=True)

    print(f"[{data_execucao}] Transformação concluída. Shape final: {df_encoded.shape}")

    # --- 3. CARGA DE DADOS PARA O S3 ---
    print(f"[{data_execucao}] Etapa 3: Carregando dados para o S3...")

    output_parquet_file = 'dados_rh_processados.parquet' # Nome do arquivo local no ambiente Colab
    bucket_name = 'datagirlsetl1008'

    s3_file_key = f'projeto_rh/funcionarios_transformados_{data_execucao}.parquet'

    try:
        # Salva o DataFrame como Parquet temporariamente no ambiente Colab
        df_encoded.to_parquet(output_parquet_file, index=False)

        s3 = boto3.client('s3') # Inicializa o cliente S3
        # Realiza o upload do arquivo Parquet temporário para o S3
        s3.upload_file(output_parquet_file, bucket_name, s3_file_key)
        print(f"[{data_execucao}] ✅ Upload de '{output_parquet_file}' para S3 em 's3://{bucket_name}/{s3_file_key}'!")
    except Exception as e:
        print(f"[{data_execucao}] ❌ Erro ao carregar para o S3: {e}")

    print(f"[{data_execucao}] Pipeline ETL concluído para esta execução.")
    return df_encoded # Retorna o DataFrame final, opcional
```

```python
import time
from datetime import datetime, timedelta

print("--- Simulação de Agendamento do Pipeline ETL ---")
print("O pipeline será executado para 3 'dias' simulados.")

# Simula 3 execuções em dias consecutivos (com um pequeno atraso para visualização)
for i in range(3):
    # Calcula a data simulada de execução no formato YYYYMMDD
    data_simulada = (datetime.now() + timedelta(days=i)).strftime("%Y%m%d")

    print(f"\n#####################################################")
    print(f"Simulando execução do dia: {data_simulada}")
    print(f"#####################################################")

    # Chama a função principal do pipeline ETL
    df_processado_atual = executar_pipeline_etl(data_simulada)

    # Opcional: Adiciona um pequeno atraso para ver a simulação passo a passo
    time.sleep(5) # Pausa por 5 segundos entre as "execuções"

print("\n--- Simulação de Agendamento Concluída ---")
print("Verifique seu bucket S3 para os arquivos gerados (com o sufixo de data).")
```
